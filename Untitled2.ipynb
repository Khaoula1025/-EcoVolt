# 1. Préparation des données temporelles


```python
import pandas as pd
df = pd.read_csv('data_smart_grid.csv')
df["DateTime"] = pd.to_datetime(df["DateTime"])
df = df.sort_values("DateTime")
df.set_index("DateTime", inplace=True)
```

### Pourquoi l’ordre temporel est critique ?

       car les donnes doivent etre sequentiel et suivre un ordre chronologique

### Pourquoi ne pas faire de shuffle ?
      car Cela introduirait des informations du futur dans le passé

# 2. Normalisation des données




```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)
```

### Pourquoi la normalisation est indispensable ?  
      chaque feature contribue équitablement

### Pourquoi MinMaxScaler pour LSTM ?
car LSTM depend des fonctions tanh et sigmoid et ces deux  fonctionnent mieux sur [0,1] ou [-1,1]

# 3. Création des séquences (Windowing)

```python
import numpy as np
def create_sequences(data, window=24):
    X, y = [], []
    for i in range(window, len(data)):
        X.append(data[i-window:i])
        y.append(data[i, 0])  # Consumption
    return np.array(X), np.array(y)

X, y = create_sequences(scaled_data, window=24)
```

### Pourquoi un LSTM attend des données en 3D ?
car l'input de LSTM est une sequence de vecteurs,une suite de vecteurs ordonnés dans le temps

### Signification de la forme : (samples, timesteps, features)
samples : nombre d’exemples

timesteps : 24 heures

features : consommation + productions énergétiques


# 4. Split Train / Test (temporel)

```python
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

### Pourquoi pas train_test_split classique ?

car il mélange les données aléatoirement et le LSTM apprend sur des séquences

# 5. Architecture LSTM

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
model = Sequential([
    LSTM(64, input_shape=(24, X.shape[2])),
    Dropout(0.2),
    Dense(1)
])

model.compile(
    optimizer="adam",
    loss="mse"
)
```

### Impact du nombre d’unités LSTM
  64 unités LSTM offrent un bon compromis. Elles sont suffisantes pour capturer les cycles journaliers et les relations non linéaires du mix énergétique, sans introduire un surapprentissage inutile



### Différence entre 1 ou plusieurs couches LSTM
Une architecture LSTM à une seule couche apprend directement les dépendances temporelles présentes dans la séquence d’entrée, ce qui est suffisant pour modéliser des séries temporelles de complexité modérée,elle à moins de paramètres, un entraînement plus rapide et un risque réduit de surapprentissage. En revanche, empiler plusieurs couches LSTM permet au modèle d’apprendre des représentations temporelles hiérarchiques : la première couche capture des motifs simples à court terme, tandis que les couches suivantes apprennent des dépendances plus abstraites et de plus long terme. Cette augmentation de capacité peut améliorer les performances sur des problèmes très complexes ou des séquences longues, mais elle nécessite davantage de données et augmente significativement le risque d’overfitting. Dans le contexte du projet EcoVolt, une seule couche LSTM constitue donc un compromis optimal entre performance, généralisation et simplicité du modèle.

# 6. Entrainement & Monitoring

```python
history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=30,
    batch_size=32
)
```

# 7. Evaluation du modèle

### Graphe Valeurs Réelles vs Valeurs Prédites

```python
import matplotlib.pyplot as plt
y_pred = model.predict(X_test)

plt.plot(y_test, label="Réel")
plt.plot(y_pred, label="Prédit")
plt.legend()
plt.show()
```

### Courbe loss vs val_loss

```python
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.legend()
plt.show()
```

### Analyse des courbes loss / val_loss
puisque les deux courbes baissent l'apprentissage est correcte
